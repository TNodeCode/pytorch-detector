```python
import os
import glob
import torch
from torchvision.io import read_image
from torchvision.transforms import v2
from PIL import Image, ImageDraw


from detectors import *

%load_ext autoreload
%autoreload 2
```

```python
root_dir = os.getcwd()
dataset_name = "spine"
resize = 512
data_root_dir = root_dir + f"\\..\\DETR\\data\\{dataset_name}"
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
model_logs = {
    "fasterrcnn_resnet50_fpn_v2": {
        "class": FasterRCNNV2Detector,
        "epochs": [80]
    },
    "fcos_resnet50_fpn": {
        "class": FCOSResNet50FPNDetector,
        "epochs": [300]
    },
    "retinanet_resnet50_fpn": {
        "class": RetinaNetResNet50FPNDetector,
        "epochs": [100]
    },
    "ssdlite320_mobilenet_v3_large": {
        "class": SSDLite320MobileNetV3LargeDetector,
        "epochs": [80]
    }
}
os.listdir(os.path.join("logs/spine", list(model_logs.keys())[0]))
print(os.path.join)
```

```python
for model_name in model_logs:
    for epoch in model_logs[model_name]["epochs"]:
        for subset in ["train", "val", "test"]:
            resume = f"logs/spine/{model_name}/{model_name}_epoch{str(epoch).zfill(4)}.pth"
            csv_name = f"logs/spine/{model_name}/{model_name}_epoch{str(epoch).zfill(4)}_{subset}.csv"
            print(model_name, epoch, subset)
            print(csv_name)
            data_dir = f"{data_root_dir}\\{subset}2017"
            detector = model_logs[model_name]["class"](
                num_classes=1,
                resume=resume,
                device=device,
            )
            inference = detector.inference()
            image_paths = glob.glob(data_dir + "/*.png")
            df = inference.get_results_df(image_paths=image_paths, resize=resize)
            df.to_csv(csv_name, index=False)
```

```python
detector = FasterRCNNV2Detector(
    num_classes=1,
    resume=f"logs/spine/fasterrcnn_resnet50_fpn_v2/fasterrcnn_resnet50_fpn_v2_epoch{str(epoch).zfill(4)}.pth",
    device=device,
)
inference = detector.inference()
```
